```python
import numpy as np
num=np.array([1,2,3,4,5])
print(num[1:4])
print(num[2:])
print(num[-3:-1])
```

```python
import numpy as np
num=np.array([1,2,3,4,5],dtype='S')
print(num.dtype) #Datatype detecting
```

```python
# Using numpy for diff datatypes
import numpy as np
num=np.array([1,2,3,4,5])
n=num.astype(bool)
new=num.astype(float)
newn=num.astype('S')
print(num.shape)
print(newn)
print(new)
print(n)
```
